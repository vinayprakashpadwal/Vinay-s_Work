```python

preference = {0: "Owner",
              1: "Customer"}

products = {1:["CocaCola",10,50],
            2:["Tomato Soup",15,60],
            3:["KitKat",20,25],
            4:["Coffee",10,30],
            5:["Lays",30,100]}
print("Welcome To Redbrick Vending Machine")
price = 0
while True:
    print("Please select 0 for Owner or 1 for Customer")
    for k,v in preference.items():
        print(f"{k} : {v}")
    select = int(input("Enter 0 or 1 for selecting your type: "))
    
    if select == 1:
        print("Below are the Remaining products")
        print("Id : Name,Remaining_Quantity,Price")
        
        for k,v in products.items():
            print(f"{k} : {v[0]},{v[1]},{v[2]}")
            
        Items = int(input("Enter the product id : "))
        Quantity = int(input("Enter quantity : "))
        products[Items][1] = (products[Items][1] - Quantity)
        price = price + (products[Items][2] * Quantity)
        print("Your Total Bill is : ",price)
        
        Message = input("Do you wish to continue? - ")
        
        if Message == "yes":
            continue;
        elif Message == "no":
            print("Thank You, Visit Again :)")
            break;
        else:
            print("Please type 'yes' or 'no'.")
            
    elif select == 0:
        Username = "vinayp"
        Password = "appleone"
        user = input("Enter username: ")
        password = input("Enter password: ")
        if(Username == user and Password == password):
            print("Remaining quantity is shown below")
            for k,v in products.items():
                print(f"{k} : {v[0]},{v[1]}")
        else:
            print("You've entered wrong username or password.")        
                
        print("Today's total sales is : ",price)
            
    else:
        print("please enter either 0 or 1 to proceed.")
    
            
   

       
        
    
    
```
